```python
#Dependancies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python

#File to load
file_to_load ='GDPdata2.csv'

#Read GDP csv file into Pandas dataframe
gdp_data= pd.read_csv(file_to_load)
gdp_data.head()
```

```python
gdp_reddata = gdp_data.drop(columns=["DGUID","VECTOR", "COORDINATE", "SCALAR_FACTOR", "SCALAR_ID", "UOM", "UOM_ID", "STATUS", "SYMBOL", "TERMINATED","DECIMALS"])
gdp_reddata.head()
```

```python
gdp_select = gdp_reddata[gdp_reddata.Prices == '2012 constant prices']
gdp_select
```

```python
gdp_select2 = gdp_select[gdp_reddata['Seasonal adjustment'] == 'Seasonally adjusted at annual rates']
gdp_select2
```

```python
gdp_select3 = gdp_select2[gdp_reddata['GEO'] == 'Canada']
gdp_select3
```

```python
gdp_select4 = gdp_select2[gdp_reddata['North American Industry Classification System (NAICS)'] == 'Information and communication technology sector [T013]']
gdp_select4a = gdp_select4.dropna()
gdp_select4a.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
reformated_gdp_select4a =gdp_select4a.pivot(index='REF_DATE', columns='North American Industry Classification System (NAICS)', values='VALUE')

# Preview that Reformatting worked
reformated_gdp_select4a.head()
```

```python
# Generate the Plot
chart_df = reformated_gdp_select4a

time= chart_df.index
indust1 = chart_df['Information and communication technology sector [T013]']


#Plot series

tick_locations = [value for value in time]

plt.figure(figsize=(20,3))
plt.plot(time,indust1, label ='Information and communication technology sector [T013]', marker = 'o', ls=':', color='red')
plt.xticks(tick_locations, time, rotation="vertical")

#Format Chart

plt.title("Annual GDP by Industry")
plt.ylabel("Dollars ($ millions)")
plt.xlabel("Time (Years)")
plt.grid(b=True, which='major', axis = 'y', color='#666666', linestyle='-')
plt.legend(loc=0)

# Save the Figure
plt.savefig("GDP by quarter 2007to 2019.png")

#Show the Figure
plt.show()
#plt.tight_layout()
```
