```python
#Dependancies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# Working with GDP Data
# File to load
file_to_load ='GDPdata2.csv'

# Read GDP csv file into Pandas dataframe
gdp_data= pd.read_csv(file_to_load)
gdp_data.head()
```

```python
# Refine dataframe to include only columns desired
gdp_reddata = gdp_data.drop(columns=["DGUID","VECTOR", "COORDINATE", "SCALAR_FACTOR", "SCALAR_ID", "UOM", "UOM_ID", "STATUS", "SYMBOL", "TERMINATED","DECIMALS"])
gdp_reddata.head()
```

```python
# Further refinement of dataset to look only at 2012 constant price index for relative measure
gdp_select = gdp_reddata[gdp_reddata.Prices == '2012 constant prices']
gdp_select.head()
```

```python
# Further refinement of dataset to look only at seasonally adjusted annual rates for relative measure
gdp_select2 = gdp_select[gdp_reddata['Seasonal adjustment'] == 'Seasonally adjusted at annual rates']
gdp_select2.head()
```

```python
# Further refinement of dataset to look only at a national level (vs at each provice - a future analysis possibility
gdp_select3 = gdp_select2[gdp_reddata['GEO'] == 'Canada']
gdp_select3.head()
```

```python
# Further refinement of dataset to look only at the desired industry's GDP - Information and communication technology sector
gdp_select4 = gdp_select3[gdp_select3['North American Industry Classification System (NAICS)'] == 'Information and communication technology sector [T013]']
gdp_select4a = gdp_select4.dropna()
gdp_select4a.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
reformated_gdp_select4a =gdp_select4a.pivot(index='REF_DATE', columns='North American Industry Classification System (NAICS)', values='VALUE')

# Preview that Reformatting worked
reformated_gdp_select4a.head()
```

```python
# Generate the Plot
chart_df = reformated_gdp_select4a

time= chart_df.index
indust1 = chart_df['Information and communication technology sector [T013]']


#Plot series

tick_locations = [value for value in time]

plt.figure(figsize=(20,3))
plt.plot(time,indust1, label ='Information and communication technology sector [T013]', marker = 'o', ls=':', color='red')
plt.xticks(tick_locations, time, rotation="vertical")

#Format Chart

plt.title("Annual GDP by Industry")
plt.ylabel("Dollars ($ millions)")
plt.xlabel("Time (Years)")
plt.grid(b=True, which='major', axis = 'y', color='#666666', linestyle='-')
plt.legend(loc=0)
 
# Save the Figure
plt.savefig("Annual Industry GDP 2007to 2019.png")

#Show the Figure
plt.show()
#plt.tight_layout()
```

```python
# Working with Main Indistry Data file used in over all analysis to obtain Industry Employment data for comparative analysis
# File to load
fpath = "main_table.csv"
industry_df = pd.read_csv(fpath)
industry_df.head()
```

```python
industry_df2 = industry_df.loc[industry_df["REF_DATE"] >= 2006]
industry_df2.head()
```

```python
# Refinement of dataset to look only at the desired industry's employment stats by selecting desired columns for analysis
industry_df3 =industry_df2[["REF_DATE","GEO","Labour force characteristics","Sex", "Age group",
          "North American Industry Classification System (NAICS)","VALUE"]]
industry_df3.reset_index(drop=True).head()
```

```python
# Refinement of dataset to look only at the desired industry's employment stats at a national level (vs Provincially)
industry_df3select = industry_df3[industry_df3['GEO'] == 'Canada']
industry_df3select.head()
```

```python
# Refinement of dataset to look only at the desired industry's employment stats
industry_df4 = industry_df3select[industry_df3select['North American Industry Classification System (NAICS)'] == 'Professional, scientific and technical services [54]']
industry_df4a = industry_df4.dropna()
industry_df4a.head()
```

```python
# Refinement of dataset to look only at the desired industry's employment stats for both sexes (vs either male or female)
industry_df5 = industry_df4a[industry_df4a['Sex'] == 'Both sexes']
industry_df5a = industry_df5.dropna()
industry_df5a.head()
```

```python
# Refinement of dataset to look only at the desired industry's employment stats for people who are employed (vs employed and unemployed together)
industry_df6 = industry_df5a[industry_df5a['Labour force characteristics'] == 'Employment']
industry_df6a = industry_df6.dropna()
industry_df6a.head()
```

```python
# Refinement of dataset to look only at the desired industry's employment stats for people who are employed (vs employed and unemployed together)
industry_df7 = industry_df6a[industry_df6a['Labour force characteristics'] == 'Employment']
industry_df7a = industry_df7.dropna()
industry_df7a.head()
```

```python
# Refinement of dataset to look only at the desired industry's employment stats for people who are in the main age bracket indicating main working years
industry_df8 = industry_df7a[industry_df7a['Age group'] == '25 to 54 years']
industry_df8a = industry_df8.dropna()
industry_df8a.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
reformated_industry_df8a = industry_df8a.pivot(index='REF_DATE', columns='North American Industry Classification System (NAICS)', values='VALUE')

# Preview that Reformatting worked
reformated_industry_df8a.head()
```

```python
# Generate the Plot
chart2_df = reformated_industry_df8a

time= chart2_df.index

indust2 = chart2_df['Professional, scientific and technical services [54]']


#Plot series

tick_locations = [value for value in time]

plt.figure(figsize=(20,3))

plt.plot(time,indust2, label ='Professional, scientific and technical services [54]', marker = 'o', ls=':', color='blue')
plt.xticks(tick_locations, time, rotation="vertical")

#Format Chart

plt.title("Industry Total Employment by Industry")
plt.ylabel("Employees (Thousands)")
plt.xlabel("Time (Years)")
plt.grid(b=True, which='major', axis = 'y', color='#666666', linestyle='-')
plt.legend(loc=0)
 
# Save the Figure
plt.savefig("Industry Employment and GDP 2007 to 2019.png")

#Show the Figure
plt.show()
plt.tight_layout()
```

```python
# Generate the dual axis Plot
chart_df = reformated_gdp_select4a
chart2_df = reformated_industry_df8a

time= chart2_df.index
indust1 = chart_df['Information and communication technology sector [T013]']
indust2 = chart2_df['Professional, scientific and technical services [54]']


fig, ax1 = plt.subplots(figsize=(20,3))
#plt.figure(figsize=(20,3))

ax2 = ax1.twinx()
ax1.plot(time, indust1, label ='Information and communication technology sector [T013]', marker = 'o', ls=':', color='red')
ax2.plot(time, indust2, label ='Professional, scientific and technical services [54]', marker = 'o', ls=':', color='blue')

ax1.set_xlabel('Time (Years)')
ax1.set_ylabel('Dollars ($ Millions)', color='red')
ax1.legend(loc=4)
ax2.set_ylabel('Employees (Thousands)', color='blue')
ax2.legend(loc=2)
plt.title("Industry Total Employment and Annual GDP by Industry")
plt.grid(b=True, which='major', axis = 'y', color='#666666', linestyle='-')



# Save the Figure
plt.savefig("Industry Employment and Annual GDP 2007 to 2019.png")

#Show the Figure
plt.show()
plt.tight_layout()

#
```
